```python
import numpy as np
import matplotlib.pyplot as plt 

np.random.seed(40)

exam_scores = np.random.rand(20)

exam_scores_final = exam_scores * 100 


plt.hist(exam_scores_final, bins=20, color= 'b')
plt.xlabel('Exam scores')
plt.ylabel('Frequency')      
plt.title('Frequency of Exam scores')
plt.show()
```
